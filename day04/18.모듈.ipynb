# 1. 모듈

* 클래스 또는 함수를 모아 놓은 파일
* 모든 파이썬 파일(.py)는 모듈
```
# 모든 모듈의 전체 함수 가져오기
import 모듈명
# 사용
모듈명.함수명()
```

````python
!python --version
````

````python
import mytotal

tot = mytotal.sum(1,100)
print(tot)
````

```
# 특정 함수만 골라서 가져오기

from 모듈명 import 함수1, 함수2, ...

# 사용
함수명()
```

````python
from mytotal import sum

tot = sum(1,200)
print(tot)
````

```
# 모듈의 전체 함수 가져오기
from 모듈명 import *

# 사용
함수명()
```

````python
from mytotal import *

tot = sum(1, 100)
print(tot)
````

```
# 모듈 별명 붙이기
import 모듈명 as 별명

# 사용
별명.함수명()
```

````python
import mytotal as mt

tot = mt.sum(1,100)
print(tot)
````
