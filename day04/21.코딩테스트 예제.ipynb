# 1. 수 정렬하기

https://www.acmicpc.net/problem/2750

```python
n = int(input())
data = list()

for i in range(n):
  #num = int(input())
  #data.append(num)
  data.append(int(input()))

data.sort()

for i in range(len(data)):
  print(data[i])
```

# 2. 음계

https://www.acmicpc.net/problem/2920

```python
num = list(map(int, input().split(' ')))

ascending = True
descending = True

for i in range(1,8):
  if num[i] > num[i-1]:
    descending = False
  elif num[i] < num[i-1]:
    ascending = False

if ascending :
  print("ascending")
elif descending:
  print("descending")
else:
  print('mixed')
```

# 3. 나이순정렬

https://www.acmicpc.net/problem/10814

```python
n = int(input())

array = []

# 변수부분에 언더바를 적어주면, 변수 없이 반복문 실행
for _ in range(n):
  data = input().split(' ')
  array.append((int(data[0]), data[1]))

print(array)

array = sorted(array, key=lambda x : x[0])

print(array)

for i in array:
  print(i[0], i[1])
```
