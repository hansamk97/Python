# 1. 람다함수

* 단일문(한줄)으로 표현되는 익명함수
* 코드상에서 한번만 사용되는 기능이 있을때, 굳이 함수로 만들지 않고 일회성으로 사용
* 함수 또한 클래스를 통해 생성된 객체

```python
def square1(x):
  return x ** 2
```

```python
print(square1(4))
```

```python
lambda x : x ** 2
```

```python
# 람다 함수 사용법 1
square2 = lambda x : x ** 2
```

```python
print(square2(4))
```

```python
# 람다 함수 사용법2
(lambda x : x ** 2)(4)
```

```python
data = ['watermelon   ', '   apple', '  banana  ', '    melon123']
```

```python
# sorted() : 정렬하는 함수
print(sorted(data))
```

```python
def myData(str):
  return len(str.strip()) # strip() : 문자열 공백 제거
```

```python
print(myData(' apple '))
```

```python
print(sorted(data, key=myData))
```

```python
# 람다 함수 사용법3
print( sorted(data, key=lambda str: len(str.strip())) )
```
