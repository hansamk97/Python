# 1. 스페셜 메소드(Special Method)

* __로 시작해서 __로 끝나는 특수함수
* 해당 메소드들을 구현하면 객체에 여러가지 파이썬 내장 함수나 연산자에 원하는 기능을 적용 가능
* 오버라이딩 가능한 함수 목록은 아래 링크에서 확인할 수 있다.
(https://docs.python.org/3/reference/datamodel.html)  

```python
class Sentence:
  def __init__(self, s):
    self.words = s.split()

  def __len__(self):  # len() 함수를 위한 스페셜 메소드
    return len(self.words)

s = Sentence('Somtimes bad things happen to good people')
```

```python
# len() : 길이값을 return 파이썬 내장 함수
print(len('Hello'))
print(len([1,2,3,4,5]))
print(s.words)
print(len(s.words))
print(len(s))
```

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y
  def __str__(self):  # str() 스페셜 메소드 기능 오버라이딩
    return ('{},{}'.format(self.x, self.y))
```

```python
p1 = Point(3,4)
print(p1)
print(str(p1))    # str(클래스객체) : object객체,
                  # str이 생략이 됐다.
                  # str에 다른 기능을 override
```
