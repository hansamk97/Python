### 1-2. 매개변수(Parameter)가 있는 함수
```
def 함수명(변수1, 변수2, ...):
  함수가 호출되면 실행할 문장
  ...
```

```
함수명(값1, 값2, ...)  # 호출  
```

````python
def func2(num):
  print('입력받은 숫자 : %d' % num)
````

````python
func2(10)
func2(20)
````

````python
# func3(1, 10) -> 1부터 10까지의 합 : 55
# func3(1, 100) -> 1부터 100까지의 합 : 5050
def func3(start, end):
  sum = 0
  for i in range(start, end+1, 1):
    sum += i
  print("%d부터 %d까지의 합 : %d" % (start, end, sum))
````

````python
func3(1,10)
func3(1,100)
func3(5,100)
````

### 1-5. 매개변수의 개수가 가변적인 함수
```
def 함수명(*변수):
  함수가 호출되면 실행할 문장
  ...
  return 값
```
```
변수 = 함수명(값1)  # 호출
변수 = 함수명(값1, 값2, ...) # 호출
```
> 매개변수는 함수안에서 튜플형태로 사용

````python
def func7(*args):
  print(args)
  result = 0
  for i in args:
    result += i;
  return result
````

````python
print(func7(10))
print(func7(10,20,30,40,50,70))
````

### 1-6. 매개변수를 딕셔너리 구조로 전달받는 함수
```
def 함수명(**변수):
  함수가 호출되면 실행할 문장
  ...
  return 값
```
```
변수 = 함수명(변수1=값1, 변수2=값2, ...) # 호출
```

````python
def func8(**args):
  return args
````

````python
user = func8(userid='admin', name='김자바',age=20)
print(user)
````

### 1-7. 리턴값이 여러개인 함수
```
def 함수명(변수1, 변수2, ...):
  함수가 호출되면 실행할 문장
  ...
  return 값1, 값2, ..
```
```
변수1, 변수2, ... = 함수명(값1, 값2, ...) # 호출
```

````python
def func9(num1, num2):
  return num1+num2, num1-num2, num1*num2, num1/num2
````

````python
result1, result2, result3, result4 = func9(10,3)
print(result1,result2,result3,result4)
````

````python
_,_,result1,_ = func9(10,3)
print(result1)
````

# 문제
데이터 90,80,50,100,70을 매개변수로 전달해서 최대값, 최소값을 출력하는 함수를 만들어 보자
(단, return을 사용, 내장함수 사용하지 않고, 최대값, 최소값을 한번에 return하는 함수를 만들어보자)

````python
def maxmin(*args):
  min,max = args[0], args[0]

  for i in args:
    if max < i:
      max = i

    if min > i:
      min = i

  return max,min
````

````python
max, min = maxmin(90,80,50,100,70)
print('최대값 : ',max)
print('최소값 : ',min)
````
