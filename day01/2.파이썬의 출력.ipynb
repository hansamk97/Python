# 2. 출력 서식 지정하기
1. %d(정수), %f(실수), %s(문자열)를 사용하여 출력 서식을 지정
2. 서식을 제외한 문자는 입력한 그대로 출력
3. 서식을 지정해 출력할 데이터가 2개 이상일 경우 ()로 데이터를 묶어야 한다.


```python
print('숫자 : 100')
print('숫자 : %d' % 100)  # 정수 100이 출력
print('숫자 : %d' % 100.12)

print('실수 : %f' % 100.12)   # 소수점 아래 6자리까지 표시

print('문자열 : %s' % 'Python')
print('문자열 : %s' % 10)     # 문자열

# 5 + 3 = 8
# 출력 서식을 이용하고 결과는 실제 덧셈 결과를 출력
print('5 + 3 = 8')  # X
print('%d + %d = %d' % (5, 3, 5+3))
```

```python
print('%5d' % 100) # 다섯자리 정수를 만든다. 앞에 두 자리는 빈공간
```

```python
print('%5d' % 100)
print('%5d' % 10000)
print('%05d' % 100)
```

```python
print('%6.2f' % 123.45)     # .포함 6자리, 6자리인데 2자리 소숫점
print('%6.2f' % 123.456)    # 자릿수 넘어가면 반올림
print('%6.2f' % 123.4)      # 남은 자리는 2으로 채움
```

```python
print('%5s' % 'abc')
```

```python
# 5 / 3 = ?
print('%d / %d = %d' % (5,3,5/3)) # 정수로 결과 확인
print('%d / %d = %f' % (5,3,5/3)) # 실수로 결과 확인
```

# 3. format 함수와 출력 인덱스 사용


```python
# 인덱스 : 0부터 순서대로 진행되는 정수
print('{0} {1}'.format('김자바','이파이'))
print('{0}{1}'.format('김자바','이파이'))
print('{1} {0}'.format('김자바','이파이'))
```

```python
# 인덱스 0번에다가 3글자이면서 String,
# 인덱스 1번에다가 3자리이면서 정수, 빈 자리는 0으로 채우기
print('{0:3s}님은 {1:03d}살입니다.'.format('자바학생',20))
print('{0}님은 {1:6.2f}cm 입니다.'.format('자바학생',160.5))
```
