# 1. 딕셔너리(Dictionary)

* 대응관계를 나타내는 자료형으로 key와 value라는 것을 한쌍으로 갖는 형태
* 하나의 딕셔너리의 키값은 중복될 수 없음
* 하나의 딕셔너리의 값은 중복될 수 있음

### 1-1. 딕셔너리 만들기

```python
dic1 = {}
print(dic1)

dic2 = {'no':1, 'userid':'apple','name':'자바학생','hp':'010-1234-5678'}
print(dic2)
```

### 1-2. 키로 값 찾기

```python
dic2 = {'no':1,'userid':'apple','name':'김자바','hp':'010-12345678'}

print(dic2['no'])
print(dic2['userid'])
print(dic2['name'])
print(dic2['hp'])
```

### 1-3. 데이터 추가 및 삭제

```python
dic3 = {1:'admin'}
print(dic3)

dic3[100] = '김자바'
print(dic3)

dic3[50] = '김자바'
print(dic3)

dic3[1] = 'user'
print(dic3)

del dic3[50]
print(dic3)

print(type(dic3)) # <class 'dict'>
```

### 1-4. 딕셔너리 함수

```python
dic2 = {'no':1,'userid':'apple','name':'김자바','hp':'010-12345678'}

# keys() : key리스트를 반환
print(dic2.keys())

# values() : value 리스트를 반환
print(dic2.values())

# items() : key와 value를 한 쌍으로 묶는 튜플을 반환
print(dic2.items())

# get() : key를 이용해서 value를 반환
print(dic2['userid'])   # apple
# print(dic2['age'])      # KeyError: 'age'
print(dic2.get('userid'))
print(dic2.get('age'))  # None
print(dic2.get('age', '알수없음'))

# in : key가 딕셔너리 안에 있는지를 확인
print('name' in dic2)   # True
print('age' in dic2)    # False
```
