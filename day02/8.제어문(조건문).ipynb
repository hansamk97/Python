### 2-2. if~else문

```
if조건식(결과가 True 또는 False가 나오는 문장 ) :
 조건식이 True인 경우 실행할 문장
else :
 조건식이 False인 경우 실행할 문장
```


````python
# 나이를 입력받아오기
# 19살 초과이면, "성인입니다."
# 그렇지 않으면 "미성년입니다"
# 조건 상관없이 : 입력한 나이 : OO
age = int(input('나이를 입력하세요 : '))

if age > 19 :
  print('성인입니다')
else :
  print('미성년 입니다')
print('입력한 나이 : ', age)
````

````python
# 나이를 입력받아오기
# 19살 초과이면, "성인입니다.", "넷플릭스 성인영화를 감살할 수 있습니다."
# 그렇지 않으면 "미성년입니다", "디즈니를 보세요"
# 조건 상관없이 : 입력한 나이 : OO

age = int(input('나이를 입력하세요 : '))

if age > 19 :
  print('성인입니다')
  print('넷플릭스 성인영화를 감살할 수 있습니다.')
else :
  print('미성년 입니다')
  print('디즈니를 보세요')
print('입력한 나이 : ', age)
````

### 2-3. if ~ elif ~ else(필수X)문

```
if조건식(결과가 True 또는 False가 나오는 문장 ) :
 조건식이 True인 경우 실행할 문장
elif 조건식2:
 조건식2가 True인 경우 실행할 문장
elif 조건식3:
 조건식3가 True인 경우 실행할 문장
...
else :
 조건식이 False인 경우 실행할 문장
```

````python
# 나이를 입력받아오기
# 19살 초과이면, "성인입니다."
# 14살 초과이면, "청소년입니다"
# 6살 초과이면, "어린이입니다"
# 0살 초과이면, "유아입니다."
# 그렇지 않으면 "입력이 잘못되었습니다"
# 조건 상관없이 : 입력한 나이 : OO
age = int(input('나이를 입력해주세요'))

if age > 19 :
  print('성인입니다.')
elif age > 14 :
  print('청소년입니다.')
elif age > 6 :
  print('어린이입니다.')
elif age > 0 :
  print('유아입니다.')
else :
  print('잘못입력하였습니다')
print('입력한 나이 : ', age)
````

### 문제
* 국어점수, 영어점수, 수학점수를 입력받아 총점, 평균, 학점을 구하는 프로그램 작성
(단, 학점은 평균 기준이며 90점 이상은 A, 80점 이상은 B, 70점 이상은 C, 60점 이상은 D, 나머지 F)




````python
kor, eng, math = input('국어, 영어, 수학점수를 입력하세요. ').split(' ')
tot = int(kor) + int(eng) + int(math)
avg = tot / 3

hak = ''

if avg >= 90:
  hak='A'
elif avg >= 80:
  hak='B'
elif avg >= 70:
  hak='C'
elif avg >= 60:
  hak='D'
else :
  hak = 'F'

print('총점 : ', tot)
print('평균 : ', avg)
print('학점 : ', hak)
````

### 논리 연산자(AND)
* 양쪽 두 식의 값이 모두 True인 경우 결과가 True가 되는 연산자

| A | B | 결과 |
|-----|-----|-----|
|True|True|True|
|True|False|False|
|False|True|False|
|False|False|False|


````python
# 로그인 프로그램 작성
# 아이디, 비밀번호 각각 입력 받은 후
# 아이디는 admin, 비밀번호는 1234가 일치하면
# '로그인 되었습니다'
# 그렇지 않으면 '아이디 또는 비밀번호를 확인하세요'

userid = input('아이디를 입력해주세요')
userpw = input('패스워드를 입력해주세요')

if userid == 'admin' and userpw == '1234':
 print('로그인 되었습니다')
else:
 print('아이디 또는 비밀번호를 확인하세요')
````

### 논리연산자(OR)
* 양쪽 두 식의 값이 하나라도 True인 경우 결과가 True가 되는 연산자

| A | B | 결과 |
|-----|-----|-----|
|True|True|True|
|True|False|True|
|False|True|True|
|False|False|False|

````python
# 주민번호에 의한 성별 판별
# 001011-3068518 주민번호를 입력('-'없이 입력해주세요)
# 7번째 자리수가 1 or 3 이면 남자 출력
# 7번째 자리수가 2 or 4 이면 여자 출력
# 그렇지 않으면, '잘못입력했습니다'

jumintNo = input('주민번호 13자리를 -없이 입력하세요')

if jumintNo[6] == '1' or jumintNo[6] == '3':
  print('남자')
elif jumintNo[6] == '2' or jumintNo[6] == '4':
  print('여자')
else:
  print('잘못입력하였습니다.')
````
